# Step 4: Data preprocessing part-II

```python
# seperating independent & dependent variables

x = dataset.drop(['Taxable_Income'], axis = 1)
```

```python
x = pd.get_dummies(x, drop_first=True)
```

```python
x.head()
```

```python
y = pd.DataFrame(dataset['Taxable_Income'])
y.head()
```

```python
for i in y.Taxable_Income:
    if i<= 30000:
        y['Taxable_Income'] = y['Taxable_Income'].replace(i,'Risky')
    else:
        y['Taxable_Income'] = y['Taxable_Income'].replace(i,'Good')
```

```python
y.head()
```

```python
# Checking for data imbalnce condition..

plt.figure(figsize=(10,6))
ax = sns.countplot(x="Taxable_Income", data=y)
```

 ### Lets carry out model building with train_test split with parameter stratify..

```python
# Splitting data into train & test

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, stratify =y, random_state = 101)
```

```python
# data balancing using stratify in train_test split..cross checking

y_train.value_counts(normalize=True)
```

```python
# data balancing using stratify in train_test split..cross checking

y_test.value_counts(normalize=True)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler

st_x= StandardScaler()  
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test)
```

### Hyperparameter tuning

 #### Changing max_depth to improve model accuracy

```python
train_accuracy = []
test_accuracy = []
for max_depth in range(1,11):
    model1 = DecisionTreeClassifier(criterion='entropy', max_depth= max_depth, random_state= 10)
    model1.fit(x_train, y_train)
    train_accuracy.append(model1.score(x_train, y_train))
    test_accuracy.append(model1.score(x_test, y_test))
```

```python
frame = pd.DataFrame({'max_depth':range(1,11),'train_acc':train_accuracy, 'test_acc':test_accuracy})
```

```python
frame
```

```python
# visulizing accuracy graph to look over max_depth
labels =['train_acc','test_acc']
fig = plt.figure(figsize=(15,6))
ax = fig.add_subplot()
plt.plot(frame['max_depth'], frame['train_acc'], marker = 'o')
plt.plot(frame['max_depth'], frame['test_acc'], marker = 'o')
ax.set_xticks(np.arange(1,11))
plt.legend(labels)
plt.xlabel('max_depth of the tree')
plt.ylabel('performance')
plt.show()
```

 # 2. Building Decision Tree suing CART i.e crieterian = Gini

```python
def grid_search_cv(x_train, y_train,cv =3):

    from sklearn.model_selection import GridSearchCV
    from sklearn.tree import DecisionTreeClassifier

    model2 =DecisionTreeClassifier()
    param_grid= {'criterion':['gini','entropy'],'splitter':['best'],'max_depth':[i for i in range(1,6)],\
                 'min_samples_split':[i for i in range(2,4)],\
                 'min_samples_leaf':[1,2,3],\
                 'max_features':[i for i in range(5,11)],'random_state':[10]}


    grid_search = GridSearchCV(estimator=model2, param_grid =param_grid,\
                               verbose=0, n_jobs=-1, scoring='accuracy', cv = 3)
    
    result = grid_search.fit(x_train, y_train)
    best_parameters =result.best_params_
    best_score = grid_search.best_score_
    
    print('best_param', best_parameters)
    print()
    print('best_score', best_score)
```

```python
grid_search_cv(x_train, y_train)
```

 ### model building:

```python
class GiniModel:

    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    '''
    i want to print class variables names as string formatted names to print
    statement that is which dataset user has provided.
    
    here i created two function 
    i) variablename() which will return string formated name of variable
    ii) variable_str_list() which will gives us a list within list of class variable name string
    '''
    # below function returns a class variable names into string format
    def variablename(var):
        import itertools
        return [tpl[0] for tpl in filter(lambda x: var is x[1], globals().items())]
    
    # this function will create list of string formatted class varable names
    def variable_str_list(self):
        self.list1 = []
        self.list1.append(variablename(self.x))
        self.list1.append(variablename(self.y))
        
    def run_model(self):
        
        self.variable_str_list()
        print('You are running model using {0} & {1}'.format(self.list1[0][0], self.list1[1][0]))
        
        print()
        
        from sklearn.tree import DecisionTreeClassifier
        from sklearn.metrics import classification_report, confusion_matrix

        gini_model = DecisionTreeClassifier(criterion='gini', random_state=10,\
                                            max_depth = 2, max_features= 5,min_samples_leaf = 2, min_samples_split = 2, splitter ='best')

        # model building
        model = gini_model.fit(self.x, self.y)
        print('model', model)
        print('-'*100)
        print()
        
        # predicting model results:
        y_pred = model.predict(self.x)

        # model evalution:
        model_score = model.score(self.x, self.y)
        print('model_score:',model_score)
        print('-'*100)
        print()
        
        model_features = model.feature_importances_
        print('model_features:', model_features)
        print('-'*100)
        print()

        # confusion matrix
        print('CONFUSION MATRIX WITH TRAINING DATASETS:')
        cm = confusion_matrix(self.y, y_pred)
        print(cm)
        print('-'*100)
        print()

        # TREE PLOT 
        print('TREE DIAGRAM:')
        plot_tree(model)
        print('-'*100)
        print()
        # feature_importance
        print('FEATURE IMPORATNCE:')
        feature_importance(model)
    
```

```python
train_data = GiniModel(x_train, y_train)
train_data.run_model()
```

```python
test_data = GiniModel(x_test, y_test)
test_data.run_model()
```
